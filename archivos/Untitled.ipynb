```python
import requests
import datetime
import os

# Configuración de la API 
API_URL = "https://metrobus-gtfs.sinopticoplus.com/gtfs-api/partnerValidation"
CREDENCIALES = {
    "usuario":"misael.cardenas.reyes",
    "senha":"@0xY%7xF"
}

def obtener_url_static():
  
    print(" Conectando a la API del Metrobús ")
    
    try:
        # 1. credenciales
        response = requests.post(API_URL, json=CREDENCIALES)
        response.raise_for_status()  # Verifica errores HTTP
        
        # 2.respuesta
        datos_respuesta = response.json()
        url_static = datos_respuesta.get('urlStatic')
        
        if url_static:
            print("\n URL Static encontrada:")
            print(url_static)
            return url_static
        else:
            print("\n Error: No se encontró urlStatic en la respuesta")
            print("Respuesta completa recibida:", datos_respuesta)
            return None
            
    except requests.exceptions.RequestException as e:
        print(f"\n Error de conexión: {str(e)}")
        return None
    except ValueError as e:
        print("\n Error: La respuesta no es un JSON válido")
        return None

if __name__ == "__main__":
    url = obtener_url_static()
    
    if url:
        fecha_actual = datetime.datetime.now()
        fecha = fecha_actual .strftime("%Y%m%d")
        hora = fecha_actual.strftime("%H_%M_%S")
        ruta_descarga = "/home/jovyan/work/descargas/" 
        nombre_archivo = f"Metrobus_GTFS_ESTATICO_{fecha}_{hora}.zip"
        ruta_completa = os.path.join(ruta_descarga, nombre_archivo)
        try:
            response = requests.get(url, stream=True)
            response.raise_for_status()  # Verifica errores HTTP
            with open(ruta_completa, "wb") as file:
                for chunk in response.iter_content(chunk_size=8192):
                    file.write(chunk)                                           
            print("(Metrobus_GTFS_ESTATICO.rar)")
        except Exception as e:
            print(f" Error {e}")
    else:
        print("\nNo se pudo obtener la URL static")
        
```
